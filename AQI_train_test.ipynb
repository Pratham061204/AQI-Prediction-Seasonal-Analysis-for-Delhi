Outliers detection and handling

Removing outliers 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def handle_outliers_iqr(data, feature):
   
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1  

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data = data[(data[feature] >= lower_bound) & (data[feature] <= upper_bound)]

    return data
```

```python
pollutants = ['Particulate Matter', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene']

for pollutant in pollutants:
    # Create a copy for visualization before outlier handling
    original_data = delhi_data.copy()

    # Handle outliers
    delhi_data = handle_outliers_iqr(delhi_data, pollutant)

    # Plot original vs. after outlier handling side-by-side
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    sns.boxplot(data=original_data, x=pollutant, color="skyblue")
    plt.title(f"{pollutant} - Before Outlier Handling")

    plt.subplot(1, 2, 2)
    sns.boxplot(data=delhi_data, x=pollutant, color="lightgreen")
    plt.title(f"{pollutant} - After Outlier Handling")

    plt.tight_layout()
    plt.show()
```

Lasso and Ridge

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

pollutants = ['Particulate Matter', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'NOx']
X = delhi_data[pollutants]
y = delhi_data['AQI']

lasso_model = Lasso(alpha=1.0)
ridge_model = Ridge(alpha=1.0)

def evaluate_model(model, X, y, test_size=0.3):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

    print(f"\n{model.__class__.__name__} Metrics on 70-30 Split:")
    print(f"RÂ² Score: {r2:.4f}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}\n")

evaluate_model(lasso_model, X, y)

evaluate_model(ridge_model, X, y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

pollutants = ['Particulate Matter', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'NOx']
X = delhi_data[pollutants]
y = delhi_data['AQI']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

best_model = None
best_score = -np.inf
best_params = {}

for n in param_grid['n_estimators']:
    for d in param_grid['max_depth']:
        for split in param_grid['min_samples_split']:
            for leaf in param_grid['min_samples_leaf']:
                model = RandomForestRegressor(
                    n_estimators=n,
                    max_depth=d,
                    min_samples_split=split,
                    min_samples_leaf=leaf,
                    random_state=0
                )
                model.fit(X_train, y_train)
                preds = model.predict(X_train)
                r2_train = r2_score(y_train, preds)

                if r2_train > best_score:
                    best_score = r2_train
                    best_model = model
                    best_params = {
                        'n_estimators': n,
                        'max_depth': d,
                        'min_samples_split': split,
                        'min_samples_leaf': leaf
                    }

y_pred = best_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("Best Hyperparameters (from train set):", best_params)
print("\nEvaluation on 70-30 Test Set:")
print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}")
```

AdaBoost Regressor with Hyperparameter Tuning and GridSearchCV

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

pollutants = ['Particulate Matter', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'NOx']
X = delhi_data[pollutants]
y = delhi_data['AQI']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

pipeline = make_pipeline(StandardScaler(), AdaBoostRegressor(random_state=42))

param_grid = {
    'adaboostregressor__n_estimators': [50, 100, 200],
    'adaboostregressor__learning_rate': [0.01, 0.1, 1.0],
    'adaboostregressor__loss': ['linear', 'square', 'exponential']
}

grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
print("Best Hyperparameters:", grid_search.best_params_)

y_pred = best_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("\nEvaluation on 70-30 Test Split:")
print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}")
```

Adaboost with RandomizedSearchCV

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

pollutants = ['Particulate Matter', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'NOx']
X = delhi_data[pollutants]
y = delhi_data['AQI']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

base_estimator = DecisionTreeRegressor(max_depth=5, random_state=42)

pipeline = make_pipeline(StandardScaler(), AdaBoostRegressor(estimator=base_estimator, random_state=42))

param_dist = {
    'adaboostregressor__n_estimators': np.arange(50, 300, 50),
    'adaboostregressor__learning_rate': np.linspace(0.01, 1.5, 10),
    'adaboostregressor__estimator__max_depth': [3, 5, 7, 10],
    'adaboostregressor__loss': ['linear', 'square', 'exponential']
}

random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=20,
    cv=5,  
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

random_search.fit(X_train, y_train)

best_model = random_search.best_estimator_
print("Best Hyperparameters:", random_search.best_params_)

y_pred = best_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("\nEvaluation on 70-30 Test Split:")
print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}")
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV, ShuffleSplit
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import numpy as np

pollutants = ['Particulate Matter', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'NOx']
X = delhi_data[pollutants]
y = delhi_data['AQI']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

base_estimator = DecisionTreeRegressor(max_depth=5, random_state=42)

pipeline = make_pipeline(StandardScaler(), AdaBoostRegressor(estimator=base_estimator, random_state=42))

param_dist = {
    'adaboostregressor__n_estimators': np.arange(50, 300, 50),
    'adaboostregressor__learning_rate': np.linspace(0.01, 1.5, 10),
    'adaboostregressor__estimator__max_depth': [3, 5, 7, 10],
    'adaboostregressor__loss': ['linear', 'square', 'exponential']
}

cv_split = ShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=20,
    cv=cv_split,
    scoring='r2',
    n_jobs=-1,
    verbose=2,
    random_state=42
)

random_search.fit(X_train, y_train)

best_model = random_search.best_estimator_
print("Best Hyperparameters:", random_search.best_params_)

y_pred = best_model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("\nEvaluation on 30% Test Set:")
print(f"RÂ² Score: {r2:.4f}")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAPE: {mape:.4f}%")
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np

pollutants = ['Particulate Matter', 'NO', 'NO2', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'NOx']
X = delhi_data[pollutants]
y = delhi_data['AQI']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model_catboost = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.15,
    depth=6,
    random_state=42,
    verbose=0
)

model_catboost.fit(X_train_scaled, y_train)

y_pred = model_catboost.predict(X_test_scaled)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

print("CatBoost Regressor - 70-30 Train-Test Split Results:")
print(f"RÂ² Score: {r2:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}%")
```
